Decision Tree Algorithm (Task6)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from google.colab import files
import io
up=files.upload()
```

Load the data

```python
data=pd.read_csv(io.BytesIO(up['Iris (1).csv']))
```

```python
data.head()
```

```python
data.describe()
```

Check for missing values

```python
data.isnull().sum()
```

```python
#There's no missing values
```

Train Test Split

```python
from sklearn.model_selection import train_test_split


###
#FEATURES
features=data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]  
#TARGET
target=data['Species']  


###
# Split dataset into training set and test set
#we will use 20% of the data for testing
train_xs, test_xs, train_ys, test_ys = train_test_split(features, target, test_size=0.2,random_state=42) 


###
print(f'train_xs shape : {train_xs.shape}')
print(f'test_xs shape : {test_xs.shape}')
print(f'train_ys shape : {train_ys.shape}')
print(f'test_ys shape : {test_ys.shape}')
```

Hyperparameter Tuning using GridSearchCV

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
criterion = ['gini', 'entropy']
param_grid = {'criterion': criterion,
               'max_depth': [1,2,3,4,5,6,7,8],
              }
print(param_grid)
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator = model, param_grid = param_grid, cv = 3)
grid.fit(train_xs, train_ys)
```

```python
grid.best_params_
```

```python
print (f'Train Accuracy - : {grid.score(train_xs,train_ys):.3f}')
print (f'Test Accuracy - : {grid.score(test_xs,test_ys):.3f}')
```

Build the Model

```python
model=grid.best_estimator_
#Train the model
model.fit(train_xs,train_ys)
```

Compare predictions and actual categories

```python
predictions=model.predict(test_xs)
comp=pd.DataFrame({'predictions':predictions,'actual_categories':test_ys})
comp.head()
```

Accuracy of the model

```python
from sklearn import metrics
print(metrics.accuracy_score(test_ys,predictions))
```

Confusion Matrix


```python
from sklearn.metrics import confusion_matrix


confusionMatrix=confusion_matrix(test_ys,predictions)


sns.heatmap(confusionMatrix,annot=True)

plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
```

Plot the Decision Tree

```python
import graphviz
from sklearn import tree
txt = tree.export_graphviz(model, out_file=None, filled=True)
graph = graphviz.Source(txt, format="png") 
graph
```
